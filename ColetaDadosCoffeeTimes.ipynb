```python
climas = pd.read_csv('/content/drive/MyDrive/dadosClimaticos_320.csv')
climas['data'] = climas['data'].str.replace('/', '-', regex=True)
climas = climas.sort_index(ascending=False).reset_index(drop=True)
```

```python
climas.head(20)
```

```python
climas['tMin'] = climas['tMin'].rolling(8).mean()
climas['tMed'] = climas['tMed'].rolling(8).mean()
climas['tMax'] = climas['tMax'].rolling(8).mean()
```

```python
climas.head(21)
```

```python
dates = []
for i in range(len(climas)):
  u = str(climas.iloc[i, 0])
  u = u.split('-')
  u1 = u[0]
  u2 = u[1]
  u3 = u[2]
  u = u3 + '-' + u2 + '-' + u1
  dates.append(u)
```

```python
climas['data'] = dates
```

```python
climas['data'] = pd.to_datetime(climas['data'])
df['data'] = pd.to_datetime(df['data'])
```

```python
df_base = pd.merge(left=climas, right=df, left_on='data', right_on='data')
```
